```python
#Check the columns unique values and drop such columns with its value as 1
unique_1_col=[]
for i in df.columns :
    if df[i].nunique() ==1:
        unique_1_col.append(i)
    else:
        pass
df.drop(unique_1_col, axis=1, inplace = True)
print("\n The following columns are dropped from the dataset as their unique value is 1. (i.e)It has no variance in the model\n",unique_1_col)
```
